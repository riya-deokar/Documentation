# Creating an OPE Textbook

Creating an OPE textbook Documentation

## Steps

1. Create an empty repsitory
2. Download [OPEEFFORT tools repository](https://github.com/OPEFFORT/tools) 
```shell

cd tools-main

```
3. Generate an SSH key, using the email associated with your github account

```shell

ssh-keygen -t ed25519 rsa -b 4096 -C quote your_email@example.com  quote

```
4. Use this command to look at the content of your key, then copy and add the SSH key to your associated github account: Settings > SSH and GPG keys

```shell

cat ~/.ssh/config/id_rsa.pub

```
5. Make sure you are able to use github commands on your linux machine, if not run the commands below

```shell

sudo apt update
sudo apt install git

```
6. Within the tools directory, clone the empty repository you created with the command below: fill in the username and repository name

```shell

git clone git@github.com:your_github_username/your_github_empty_repo.git

```
7. Install the OPE Command Line Tool by running the install.sh fil in the tools directory as shown below

```shell

./install.sh
```
After that paste the outputted export path line in your command-line
Lastly, run the ope command to ensure installation

```shell

ope

```
8. Run this command (including your own project repo name and repo url) to create a new project using the empty repository URL as shown below
You can find your repoâ€™s SSH URL when you the green <> Code button on your github repo
This will create subdirectories of books, containers, and content in your github repo

```shell

ope new_project your_project_name your_repo_url
```
9. Create a new book in your repository, replacing your_book_name

```shell

ope new_book your_book_name

```
10. Move to your book's content directory, edit the configure file to change book settings (title, author, logo, URL to publish to, and jupyter extensions enabling)
If you want to change your textbook's logo on the top left corner of your book: change the logo .png name and place a new png in the content directory
